```python
import turtle
# 多边形的顶点序列
vertices = [(2,4), (5,1), (7,1), (12,5), (12,11), (7,7), (2,8)]

# 创建一个画布，并将画笔移到第一个点
turtle.setup()
turtle.penup()
turtle.goto(vertices[0][0], vertices[0][1])
turtle.pendown()

# 画多边形的边界
for i in range(len(vertices)):
    j = (i + 1) % len(vertices)
    turtle.goto(vertices[j][0], vertices[j][1])

# 创建一个有序边表
edge_table = {}
for i in range(len(vertices)):
    j = (i + 1) % len(vertices)
    if vertices[i][1] < vertices[j][1]:
        y_min = vertices[i][1]
        x_at_y_min = vertices[i][0]
        y_max = vertices[j][1]
        slope = (vertices[j][0] - vertices[i][0]) / (vertices[j][1] - vertices[i][1])
        edge_table.setdefault(y_min, []).append((slope, y_max, x_at_y_min))
    elif vertices[i][1] > vertices[j][1]:
        y_min = vertices[j][1]
        x_at_y_min = vertices[j][0]
        y_max = vertices[i][1]
        slope = (vertices[j][0] - vertices[i][0]) / (vertices[j][1] - vertices[i][1])
        edge_table.setdefault(y_min, []).append((slope, y_max, x_at_y_min))

# 反转有序边表，避免出现负斜率
keys = sorted(edge_table.keys())
sorted_edge_table = {key: sorted(edge_table[key]) for key in keys}

# 扫描多边形并填充
for y in range(min(vertices, key=lambda v: v[1])[1], max(vertices, key=lambda v: v[1])[1] + 1):
    try:
        if y in sorted_edge_table:
            active_edges = sorted_edge_table[y]

            # 将活动边按照与扫描线交点的x坐标排序
            active_edges.sort(key=lambda e: e[2])

            # 填充扫描线上的像素
            for i in range(0, len(active_edges), 2):
                x_start, x_end = int(active_edges[i][2]), int(active_edges[i+1][2])
                turtle.penup()
                turtle.goto(x_start, y)
                turtle.pendown()
                turtle.goto(x_end, y)

            # 更新活动边中的交点x坐标
            for i in range(len(active_edges)):
                slope, y_max, x_at_y_min = active_edges[i]
                active_edges[i] = (slope, y_max, x_at_y_min + slope)

    except IndexError:
        pass

turtle.done()
```
